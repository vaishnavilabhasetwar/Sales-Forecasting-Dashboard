```python
## Parse the Order Date and Ship Date columns as proper datetime objects
# ==========================
# Convert Date Columns
# ==========================
date_columns = ["Order Date", "Ship Date"]

for col in date_columns:
    df[col] = pd.to_datetime(
        df[col],
        format="%d/%m/%Y",   # Explicit format is faster than automatic parsing
        errors="coerce"      # Invalid dates become NaT instead of raising an error
    )

# ==========================
# Validate Conversion
# ==========================
print(df[date_columns].dtypes)

# Check for invalid dates
invalid_dates = df[date_columns].isna().sum()

print("\nMissing/Invalid Dates:")
print(invalid_dates)
```

## Plot all 4 components clearly on one figure


```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# ==========================================================
# Monthly Sales Time Series
# ==========================================================

monthly_sales = (
    df.resample("ME", on="Order Date")
      .agg(Total_Sales=("Sales", "sum"))
)

# ==========================================================
# Time Series Decomposition
# ==========================================================

decomposition = seasonal_decompose(
    monthly_sales["Total_Sales"],
    model="additive",
    period=12
)

# ==========================================================
# Plot All Components
# ==========================================================

fig, axes = plt.subplots(
    nrows=4,
    ncols=1,
    figsize=(15, 12),
    sharex=True
)

# Original Sales
axes[0].plot(decomposition.observed, color="blue")
axes[0].set_title("Observed Monthly Sales", fontsize=13)
axes[0].set_ylabel("Sales")
axes[0].grid(alpha=0.3)

# Trend
axes[1].plot(decomposition.trend, color="green")
axes[1].set_title("Trend Component", fontsize=13)
axes[1].set_ylabel("Trend")
axes[1].grid(alpha=0.3)

# Seasonal
axes[2].plot(decomposition.seasonal, color="orange")
axes[2].set_title("Seasonal Component", fontsize=13)
axes[2].set_ylabel("Seasonality")
axes[2].grid(alpha=0.3)

# Residual
axes[3].plot(decomposition.resid, color="red")
axes[3].set_title("Residual (Noise) Component", fontsize=13)
axes[3].set_ylabel("Residual")
axes[3].set_xlabel("Order Date")
axes[3].grid(alpha=0.3)

plt.suptitle(
    "Time Series Decomposition of Monthly Sales",
    fontsize=16,
    fontweight="bold"
)

plt.tight_layout(rect=[0, 0, 1, 0.97])

plt.show()
```

## Overall Conclusion

The decomposition indicates that monthly sales are influenced by both a long-term growth trend and recurring seasonal effects. Since seasonality is present, forecasting models such as SARIMA or Prophet would be more suitable than models that ignore seasonal patterns.

# Check for stationarity using the Augmented Dickey-Fuller (ADF) Test — explain in plain English what stationarity means and what your test result tells you


## Import the ADF Test

```python
from statsmodels.tsa.stattools import adfuller
```

## Run the ADF Test

```python
# ==========================================================
# Augmented Dickey-Fuller (ADF) Test
# ==========================================================

adf_result = adfuller(monthly_sales["Total_Sales"])

print("Augmented Dickey-Fuller Test Results")
print("-" * 40)
print(f"ADF Statistic : {adf_result[0]:.4f}")
print(f"P-value       : {adf_result[1]:.4f}")
print(f"Lags Used     : {adf_result[2]}")
print(f"Observations  : {adf_result[3]}")

print("\nCritical Values:")

for key, value in adf_result[4].items():
    print(f"{key}: {value:.4f}")
```

## Interpret the Result Automatically

```python
alpha = 0.05

if adf_result[1] < alpha:
    print("\nResult:")
    print("The time series is STATIONARY.")
    print("The null hypothesis is rejected.")
else:
    print("\nResult:")
    print("The time series is NOT STATIONARY.")
    print("The null hypothesis cannot be rejected.")
```

## What is Stationarity?

Stationarity means that the statistical properties of a time series, such as its mean and variance, remain relatively constant over time. A stationary time series does not exhibit unpredictable changes in its behavior, making it suitable for many forecasting techniques.

## ADF Test Result

The Augmented Dickey-Fuller (ADF) Test was performed to determine whether the monthly sales time series is stationary. The test produced an ADF Statistic of -4.4161 and a p-value of 0.0003, which is less than the significance level of 0.05. Therefore, the null hypothesis of non-stationarity is rejected.

##  Observation

Observation: Since the p-value (0.0003) is less than 0.05, the monthly sales time series is stationary. This indicates that the statistical properties of the series remain relatively stable over time, making it suitable for time-series forecasting models that require stationary data.

## Apply differencing if the series is non-stationary and re-test


```python
# ==========================================================
# Apply First-Order Differencing (Only if Non-Stationary)
# ==========================================================

differenced_series = monthly_sales["Total_Sales"].diff().dropna()

# Re-run the ADF Test
from statsmodels.tsa.stattools import adfuller

adf_diff = adfuller(differenced_series)

print("ADF Statistic :", adf_diff[0])
print("P-value :", adf_diff[1])

if adf_diff[1] < 0.05:
    print("The differenced series is stationary.")
else:
    print("The differenced series is still non-stationary.")
```

## Differencing and Re-testing

Although the original monthly sales series was already stationary, first-order differencing was applied to demonstrate the standard procedure used when dealing with non-stationary time series. After differencing, the Augmented Dickey-Fuller (ADF) test was performed again.

## Observation

The differenced series produced an ADF Statistic of -8.7271 and a p-value of 3.27 × 10⁻¹⁴, which is significantly lower than the 0.05 significance level. This confirms that the differenced series is stationary, and the null hypothesis of non-stationarity is rejected.

## Conclusion

Since the original monthly sales series was already stationary (p-value = 0.0003), differencing was not required for this dataset. However, the re-test after first-order differencing confirms that the transformed series is also stationary. In practice, the original stationary series would be preferred to avoid unnecessary transformations.

### Step 3: Prepare Monthly Sales Data

```python
monthly_sales = (
    df.resample("ME", on="Order Date")
      .agg(Total_Sales=("Sales", "sum"))
)

monthly_sales.head()
```

### Step 4: Train the SARIMA Model

```python
sarima_model = SARIMAX(
    monthly_sales["Total_Sales"],
    order=(1, 0, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_results = sarima_model.fit()

print(sarima_results.summary())
```

## Why did we choose these parameters?
Non-seasonal parameters (p, d, q) = (1, 0, 1)
p = 1
Uses one previous observation to predict the next.
d = 0
Your ADF test showed the series is already stationary (p-value = 0.0003), so no differencing is needed.
q = 1
Accounts for one lag of forecast error.
## Seasonal parameters (P, D, Q, m) = (1, 1, 1, 12)
P = 1
One seasonal autoregressive term.
D = 1
Captures yearly seasonal differences.
Q = 1
One seasonal moving-average term.
m = 12
Because the data is monthly, one seasonal cycle = 12 months.

This is a widely used baseline SARIMA configuration for monthly retail sales data.

### Step 5: Forecast Next 3 Months

```python
forecast = sarima_results.get_forecast(steps=3)

forecast_mean = forecast.predicted_mean

forecast_ci = forecast.conf_int()

print(forecast_mean)
```

### Step 6: Plot Actual vs Forecast

```python
plt.figure(figsize=(14,6))

# Actual Sales
plt.plot(
    monthly_sales.index,
    monthly_sales["Total_Sales"],
    label="Actual Sales",
    linewidth=2
)

# Forecast
plt.plot(
    forecast_mean.index,
    forecast_mean,
    color="red",
    marker="o",
    linewidth=2,
    label="Forecast"
)

# Confidence Interval
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:,0],
    forecast_ci.iloc[:,1],
    color="pink",
    alpha=0.3,
    label="95% Confidence Interval"
)

plt.title("SARIMA Forecast (Next 3 Months)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Monthly Sales")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### Step 7: Forecast Table

```python
forecast_df = pd.DataFrame({
    "Forecasted Sales": forecast_mean,
    "Lower CI": forecast_ci.iloc[:,0],
    "Upper CI": forecast_ci.iloc[:,1]
})

forecast_df
```

## Create a Reusable Forecasting Function

```python
def xgboost_forecast(data, title):
    
    # Monthly Sales
    monthly = (
        data.resample("ME", on="Order Date")
            .agg(Total_Sales=("Sales", "sum"))
    )

    # Create Features
    ml = monthly.copy()

    ml["Lag_1"] = ml["Total_Sales"].shift(1)
    ml["Lag_2"] = ml["Total_Sales"].shift(2)
    ml["Lag_3"] = ml["Total_Sales"].shift(3)

    ml["Rolling_Mean_3"] = (
        ml["Total_Sales"]
        .rolling(3)
        .mean()
        .shift(1)
    )

    ml["Month"] = ml.index.month
    ml["Quarter"] = ml.index.quarter

    season_map = {
        12: "Winter", 1: "Winter", 2: "Winter",
        3: "Spring", 4: "Spring", 5: "Spring",
        6: "Summer", 7: "Summer", 8: "Summer",
        9: "Autumn", 10: "Autumn", 11: "Autumn"
    }

    ml["Season"] = ml.index.month.map(season_map)

    ml = pd.get_dummies(
        ml,
        columns=["Season"],
        drop_first=True
    )

    ml = ml.dropna()

    # Features & Target
    X = ml.drop(columns="Total_Sales")
    y = ml["Total_Sales"]

    # Train/Test Split
    split = int(len(ml) * 0.8)

    X_train = X.iloc[:split]
    X_test = X.iloc[split:]

    y_train = y.iloc[:split]
    y_test = y.iloc[split:]

    # Train Model
    model = XGBRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=3,
        random_state=42,
        objective="reg:squarederror"
    )

    model.fit(X_train, y_train)

    # Predictions
    pred = model.predict(X_test)

    # Metrics
    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2 = r2_score(y_test, pred)

    print("=" * 60)
    print(title)
    print("=" * 60)
    print(f"MAE : {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R²  : {r2:.4f}")

    # Plot
    plt.figure(figsize=(12,5))

    plt.plot(
        y_test.index,
        y_test,
        marker='o',
        label="Actual"
    )

    plt.plot(
        y_test.index,
        pred,
        marker='s',
        label="Predicted"
    )

    plt.title(title)

    plt.xlabel("Month")
    plt.ylabel("Sales")

    plt.legend()

    plt.grid(alpha=0.3)

    plt.show()

    # Forecast Next 3 Months
    sales_history = monthly["Total_Sales"].tolist()

    future_dates = pd.date_range(
        monthly.index[-1] + pd.offsets.MonthEnd(1),
        periods=3,
        freq="ME"
    )

    future_pred = []

    for d in future_dates:

        row = {
            "Lag_1":[sales_history[-1]],
            "Lag_2":[sales_history[-2]],
            "Lag_3":[sales_history[-3]],
            "Rolling_Mean_3":[np.mean(sales_history[-3:])],
            "Month":[d.month],
            "Quarter":[d.quarter],
            "Season":[season_map[d.month]]
        }

        row = pd.DataFrame(row)

        row = pd.get_dummies(row)

        row = row.reindex(
            columns=X.columns,
            fill_value=0
        )

        p = model.predict(row)[0]

        future_pred.append(p)

        sales_history.append(p)

    forecast = pd.DataFrame({
        "Month": future_dates,
        "Forecast": future_pred
    })

    print("\nNext 3-Month Forecast")
    print(forecast)

    return forecast
```

## Apply a second anomaly detection method: Z-Score based detection — flag any week where sales deviate more than 2 standard deviations from the rolling mean


```python
# ==========================================================
# Z-Score Based Anomaly Detection using Rolling Statistics
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Weekly sales
weekly_sales = (
    df.resample("W", on="Order Date")
      .agg(Total_Sales=("Sales", "sum"))
)

# Create copy
zscore_df = weekly_sales.copy()

# Rolling window (12 weeks)
window = 12

# Rolling Mean and Rolling Standard Deviation
zscore_df["Rolling_Mean"] = (
    zscore_df["Total_Sales"]
    .rolling(window=window)
    .mean()
)

zscore_df["Rolling_STD"] = (
    zscore_df["Total_Sales"]
    .rolling(window=window)
    .std()
)

# Remove initial NaN values
zscore_df = zscore_df.dropna()

# Calculate Rolling Z-Score
zscore_df["Z_Score"] = (
    zscore_df["Total_Sales"] -
    zscore_df["Rolling_Mean"]
) / zscore_df["Rolling_STD"]

# Flag anomalies (|Z| > 2)
zscore_df["Z_Anomaly"] = np.where(
    np.abs(zscore_df["Z_Score"]) > 2,
    "Anomaly",
    "Normal"
)

# Display counts
print("Z-Score Anomaly Count")
print(zscore_df["Z_Anomaly"].value_counts())

# Display anomalies
z_anomalies = zscore_df[
    zscore_df["Z_Anomaly"] == "Anomaly"
]

print("\nDetected Anomalies")
display(z_anomalies)

# ==========================================================
# Plot
# ==========================================================

plt.figure(figsize=(15,6))

plt.plot(
    zscore_df.index,
    zscore_df["Total_Sales"],
    linewidth=2,
    label="Weekly Sales"
)

plt.scatter(
    z_anomalies.index,
    z_anomalies["Total_Sales"],
    color="red",
    marker="X",
    s=180,
    edgecolors="black",
    label="Z-Score Anomalies"
)

plt.title("Weekly Sales Anomaly Detection using Rolling Z-Score")
plt.xlabel("Order Date")
plt.ylabel("Weekly Sales")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Compare: do both methods flag the same anomalies, or do they disagree? What does this tell you?


## Comparison of Anomaly Detection Methods

### Isolation Forest vs. Rolling Z-Score

| Method | Number of Anomalies Detected | Result |
|:--------|-----------------------------:|:-------|
| **Isolation Forest** | 11 | Detected multiple unusual sales weeks |
| **Rolling Z-Score (±2 SD)** | 0 | No anomalies detected |

### Observations

- The two anomaly detection methods **did not identify the same anomalies**.
- **Isolation Forest** detected **11 anomalous weeks**, including both unusually high and unusually low sales periods.
- In contrast, the **Rolling Z-Score** method did not detect any anomalies because none of the weekly sales values exceeded **±2 standard deviations** from the rolling mean.
- This difference occurs because the two methods use different approaches. Isolation Forest is a machine learning algorithm that identifies observations that are isolated from the overall data distribution, while the Rolling Z-Score is a statistical method that relies on a fixed threshold relative to the local rolling average.
- As a result, Isolation Forest was more sensitive to unusual sales patterns, whereas the Rolling Z-Score method was more conservative for this dataset.

### Conclusion

The comparison shows that different anomaly detection techniques can produce different results on the same data. Isolation Forest was more effective at identifying unusual sales behavior in this dataset, while the Rolling Z-Score method considered all observations to be within the expected statistical range. Therefore, the choice of anomaly detection method should depend on the characteristics of the data and the business objective.

## Apply K-Means Clustering to segment products into demand groups


```python
# ==========================================================
# Select Features for Clustering
# ==========================================================

features = [
    "Total_Sales",
    "Sales_Growth_Rate",
    "Sales_Volatility",
    "Average_Order_Value"
]

X = clustering_df[features]
```

## Standardize the Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

## Find the Optimal Number of Clusters (Elbow Method)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(1, 8):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    
    kmeans.fit(X_scaled)
    
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))

plt.plot(
    range(1,8),
    wcss,
    marker='o',
    linewidth=2
)

plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.grid(alpha=0.3)

plt.show()
```

## Determining the Optimal Number of Clusters using the Elbow Method

The Elbow Method was used to determine the optimal number of clusters for K-Means clustering. The method calculates the Within-Cluster Sum of Squares (WCSS) for different values of K and identifies the point where the reduction in WCSS begins to slow significantly.

From the Elbow Curve, the WCSS decreases rapidly between **K = 1** and **K = 4**, after which the curve starts to flatten. This indicates that increasing the number of clusters beyond **4** provides only a small improvement in cluster compactness.

Therefore, **K = 4** was selected as the optimal number of clusters for product demand segmentation.
